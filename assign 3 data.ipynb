```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
```

```python
# df = [['Category', 'Num'], ['Number of People who speak Siksika', 465], ['Number of People who speak Siksika as their Mother Tongue', 175], ['Number of People who speak Siksika as their most often Home Spoken Language', 35], ['Number of People who speak Siksika as their most often Work Spoken Language', 0]]

df = pd.DataFrame({ 'Category' : ['People who speak Siksika     ', 'People who speak Siksika as their Mother Tongue', 'People who speak Siksika as their most often Home Spoken Language', 'People who speak Siksika as their most often Work Spoken Language'],
                    'Number of People' : [465, 175, 35, 0] })

df
```

```python
plt.figure(figsize=(7, 5))

a = sns.barplot(data=df, y='Number of People', x='Category', orient='v')# x="Category", y="Number Of People")

a.set_xticklabels(a.get_xticklabels(), rotation=20, ha="right")
```

```python
df2 = pd.DataFrame({ 'Category' : ['People of Blackfoot Decent who live on reserve', 'People of Blackfoot Decent who live in Calgary',],
                    'Number of People' : [7565, 1705] })

df2
```

```python
plt.figure(figsize=(7, 5))

bar_colors = ['firebrick', 'purple']

a = sns.barplot(data=df2, y='Number of People', x='Category', orient='v', palette=bar_colors)# x="Category", y="Number Of People")

a.set_xticklabels(a.get_xticklabels(), rotation=15, ha="right")
```
